```python
import os
from IPython.display import HTML
from base64 import b64encode

def display_video(path):
    mp4 = open(path, 'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML("""
        <video width=400 controls>
            <source src="%s" type="video/mp4">
        </video>
    """ % data_url
    )
```

```python
from moviepy import VideoFileClip

def extract_audio(*, video_file_path: str, audio_file_path: str)-> None:
    try:
        video_clip = VideoFileClip(video_file_path)
        audio_clip = video_clip.audio
        if audio_clip is not None:
            audio_clip.write_audiofile(audio_file_path)
            audio_clip.close()
            video_clip.close()
    except Exception as e:
        raise ValueError(f'Error extracting audio: {e}')
```

```python
def word_cleaner(word: str) -> str:
    return "".join(char for char in word if char.isalnum()).lower().strip()

def query_transcript(pleeb_words: list, timestamped_transcript: list) -> list:
    transcript_words = [word for segment in timestamped_transcript for word in segment["words"]]
    detected_pleeb_words = []
    for bw in pleeb_words:
        detected_pleeb_words += [v for v in transcript_words if word_cleaner(v["text"]) == word_cleaner(bw)]
    detected_pleeb_words = sorted(detected_pleeb_words, key=lambda x: x["start"])
    return detected_pleeb_words
print(query_transcript(pleeb_words, timestamped_transcript))
```

```python
import random
from pydub import AudioSegment

meme_sounds = [
    "pleeb/sounds/bleep.mp3",
    "pleeb/sounds/bruh.mp3",
    "pleeb/sounds/huh.mp3",
    "pleeb/sounds/metal_boom.mp3",
    "pleeb/sounds/minecraft_oof.mp3",
    "pleeb/sounds/nope.mp3",
    "pleeb/sounds/screaming_sheep.mp3",
    "pleeb/sounds/windows_error.mp3",
    "pleeb/sounds/yeet.mp3",
    ]

def add_meme_sounds(og_audio_path: str, bleep_words: list) -> AudioSegment:
    original_audio = AudioSegment.from_mp3(og_audio_path)
    bleep_word_instances = query_transcript(bleep_words, timestamped_transcript)

    output = AudioSegment.empty()
    cursor = 0  # current position in the original audio

    for word in bleep_word_instances:
        start_time = int(word["start"] * 1000)
        end_time = int(word["end"] * 1000)

        # Append audio before bleep
        output += original_audio[cursor:start_time]

        # Choose random meme sound and resize to fit exact word duration
        meme_sound = AudioSegment.from_mp3(random.choice(meme_sounds))
        duration = end_time - start_time
        if len(meme_sound) > duration:
            meme_sound = meme_sound[:duration]
        else:
            meme_sound = meme_sound + AudioSegment.silent(duration - len(meme_sound))

        # Append meme sound
        output += meme_sound

        # Move cursor
        cursor = end_time

    # Append any remaining audio
    output += original_audio[cursor:]
    return output
```

```python
from moviepy import VideoFileClip, AudioFileClip, CompositeAudioClip

def pleeb(og_video_path: str, og_audio_path: str, final_video_path: str, final_audio_path: str, bleep_words: list):
    bleeped_audio = add_meme_sounds(og_audio_path, bleep_words)
    bleeped_audio.export(final_audio_path, format="mp3")

    video_clip = VideoFileClip(og_video_path)
    audio_clip = AudioFileClip(final_audio_path)

    video_clip.audio = audio_clip
    video_clip.write_videofile(
        final_video_path,
        codec="libx264",
        audio_codec="aac",
        temp_audiofile="temp-audio.m4a",
        remove_temp=True
    )
```
